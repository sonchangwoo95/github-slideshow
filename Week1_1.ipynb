스파르타 코드

```python
import cv2

img = cv2.imread('01.jpg')

print(img.shape)                # (404, 640, 3) = (높이, 너비, 채널)

# 도형 그리기
 # 1 사각형 - pt1:사각형의 왼쪽 위 x,y좌표, pt2:오른쪽 아래 x,y좌표, color BRG순서, 두께
cv2.rectangle(img, pt1=(259, 89), pt2=(380, 348), color=(255,0,0), thickness=2)
 # 2 원형 - pt1:사각형의 왼쪽 위 x,y좌표, pt2:오른쪽 아래 x,y좌표, color BRG순서, 두께
cv2.circle(img, center = (320, 220), radius = 100, color =(0,255,0), thickness=4)

# 이미지 수정하기
 # 1 자르기(crop) [y1:y2, x1:x2]
img_cropped = img[89:348, 259:380]
 # 2 크기 변경(resize) (x,y)크기
img_resized = cv2.resize(img, dsize=(512, 256))
 # 3 컬러 바꾸기
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
 # 4 회색
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)



# 수정하기 출력
cv2.imshow('cropped', img_cropped)
cv2.imshow('resized', img_resized)
cv2.imshow('rgb', img_rgb)
cv2.imshow('gray', img_gray)

# 도형 출력
cv2.imshow('result', img)       # 화면 출력
cv2.waitKey(0)                  # 아무키를 입력하기전까지 화면이 안 사라짐
```
